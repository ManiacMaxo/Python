<h3>formula</h3>

```python
from pylab import *
from numpy import *
```

```python
x = linspace(-1, 1, 500)
y = x * 2 + 1 + random.normal(size=500) / 3

a = sum((x - mean(x)) * (y - mean(y))) / sum((x - mean(x))**2)
b = mean(y) - a * mean(x)
```

```python
print('a is ', a)
print('b is ', b)
```

```python
model_x = linspace(-1, 1, 2);
scatter(x, y)
plot(model_x, model_x * a + b, 'r-', linewidth=2)
show()
```
